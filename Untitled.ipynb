```python
%matplotlib inline
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision.models import resnet18
import matplotlib.pyplot as plt
import numpy as np
import time

# Initialize model and data
model = resnet18(pretrained=True)
model.fc = nn.Linear(512, 200)  # Adjust for 200 bird classes
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training parameters
num_steps = 100
losses = []
gradients = []

# Training loop
for step in range(num_steps):
    # Training step
    optimizer.zero_grad()
    dummy_input = torch.randn(1, 3, 224, 224)  # Simulate a bird image input
    output = model(dummy_input)
    target = torch.tensor([0])  # Target class (e.g., first bird species)
    loss = nn.CrossEntropyLoss()(output, target)
    loss.backward()
    
    # Calculate gradient norm
    grad_norm = sum(p.grad.norm().item() for p in model.parameters())
    
    # Store metrics
    losses.append(loss.item())
    gradients.append(grad_norm)
    
    # Update weights
    optimizer.step()
    
    # Print progress
    print(f"Step {step}: Loss = {loss.item():.4f}, Gradient Norm = {grad_norm:.4f}")
    time.sleep(0.1)

# Show final plots
plt.figure(figsize=(15, 5))

# Loss plot
plt.subplot(1, 2, 1)
plt.plot(losses, 'b-', label='Loss')
plt.title('Training Loss')
plt.xlabel('Steps')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()

# Gradient norm plot
plt.subplot(1, 2, 2)
plt.plot(gradients, 'r-', label='Gradient Norm')
plt.title('Gradient Magnitude')
plt.xlabel('Steps')
plt.ylabel('Gradient Norm')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

print("\nTraining completed!")
print(f"Final loss: {losses[-1]:.4f}")
print(f"Final gradient norm: {gradients[-1]:.4f}")
```
